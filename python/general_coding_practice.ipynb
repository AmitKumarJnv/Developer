## Last day of the month

```python
from datetime import datetime, timedelta

def is_last_day_of_month(date: datetime) -> bool:
    next_day = date + timedelta(days=1)
    return next_day.month != date.month  # If next month is different, it's the last day

# Example usage
today = datetime.today()
print(is_last_day_of_month(today))  # True/False
```

```python
import calendar
from datetime import datetime

def is_last_day_of_month(date: datetime) -> bool:
    last_day = calendar.monthrange(date.year, date.month)[1]  # Get last day of the month
    return date.day == last_day

# Example usage
today = datetime.today()
print(is_last_day_of_month(today))  # True if today is the last day of the month
```

## Mode 

```python
from typing import List

def mode_from_list(list_of_numbers:List)->List:
    """Function to return list of mode values from a given list of numbers"""
    number_frequency = {}
    for num in list_of_numbers:
        if num in number_frequency:
            number_frequency[num] += 1
        else:
            number_frequency[num] = 1    

    mod_values = []
    max_value = max(number_frequency.values())  
    for key , value in number_frequency.items():
        if value == max_value:
            mod_values.append(key)

    return mod_values              
```

```python
mode_from_list([1,1,2,2,3,3,4,4,444])
```

```python
from collections import Counter

def mode_using_counter(list_of_numbers:List)->List:
    c = Counter(list_of_numbers)
    return c.most_common(1)[0][0]
```
