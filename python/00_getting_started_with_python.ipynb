```python
# Naming Convention for variables
## Variable names should be descriptive
## They must start with a letter or an '_' and contain letters, numbers or underscores 

## Proper var_names example
var_01 = 0
_var_02 = 5

## Improper var_names
@name = "Name" # SyntaxError: invalid syntax. Maybe you meant '==' or ':=' instead of '='?
var-name  = "name" # SyntaxError: cannot assign to expression here. Maybe you meant '==' instead of '='?
```

```python
# Type conversion

age = '25'
print(type(int(age))) # <class 'int'>
name = "name" 
print(type(int(name))) # ValueError: invalid literal for int() with base 10: 'name'
```
