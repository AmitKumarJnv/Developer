```python
"""
insert i e: Insert integer  at position .
print: Print the list.
remove e: Delete the first occurrence of integer .
append e: Insert integer  at the end of the list.
sort: Sort the list.
pop: Pop the last element from the list.
reverse: Reverse the list."""

N = int(input())
sample_list = []
for i in range(N):
    command,*variables = input().split()
    if command == "insert":
        sample_list.insert(int(variables[0]),int(variables[1]))
    elif command == "print":
        print(sample_list)
    elif command == "remove":
        sample_list.remove(int(variables[0]))
    elif command == "append":
        sample_list.append(int(variables[0]))
    elif command == "sort":
        sample_list.sort()
    elif command == "pop":
        sample_list.pop()
    elif command == "reverse":
        sample_list.reverse()
    else:
        print("wrong command")                
```

```python
def count_substring(string, sub_string):
    sub_string_len = len(sub_string)
    counter = 0
    for i in range(len(string)-sub_string_len+1):
        if string[i:sub_string_len] == sub_string:
            counter+=1
        sub_string_len+=1
    return counter
```

```python
def new_doormat_specification(n: int, m: int):
    pattern = ".|."
    # Generate only the top half
    top_half = [((2 * i + 1) * pattern).center(m, "-") for i in range(n // 2)]
    
    # Print top, middle, and reversed top
    print('\n'.join(top_half))
    print("WELCOME".center(m, "-"))
    print('\n'.join(reversed(top_half)))
```

```python
def minion_game(string):
    vowels = "AEIOU"
    kevin_score = 0
    stuart_score = 0
    length = len(string)
    
    # Loop through each character in the string
    for i in range(length):
        # The number of substrings starting at this index
        score = length - i 
        
        if string[i] in vowels:
            kevin_score += score
        else:
            stuart_score += score
            
    # Determine the winner
    if stuart_score > kevin_score:
        print(f"Stuart {stuart_score}")
    elif kevin_score > stuart_score:
        print(f"Kevin {kevin_score}")
    else:
        print("Draw")

# HackerRank handles the input like this:
if __name__ == '__main__':
    s = input()
    minion_game(s)
```

```python
def merge_the_tools(string, k):
    name_list_divided = [string[i:i+k] for i in range(0, len(string), k)]
    for chunk in name_list_divided:
        unique_chars = []
        for char in chunk:
            if char not in unique_chars:
                unique_chars.append(char)
                
        print("".join(unique_chars))

merge_the_tools('AABCAAADA', 3)
```

```python
from typing import List,Set
def final_score(input_array:List[int],A:Set[int],B:Set[int])->int:
    score = 0
    for element in input_array:
        if element in A:
            score += 1
        if element in B:
            score -= 1
    return score          


if __name__ == "__main__":
    n,m = map(int,input().split())
    n_array = list(map(int,input().split()))
    A = set(map(int,input().split()))
    B = set(map(int,input().split()))
    print(f"n:{n},m:{m},n_array:{n_array},A:{A},B:{B}")
    print(final_score(n_array,A,B))
```

```python
from collections import Counter

lst = [1, 1, 2, 3, 4, 5]
target_set = {1, 4, 5}

# Count elements in the list
counter = Counter(lst)          # Counter({1:2, 2:1, 3:1, 4:1, 5:1})

# Intersection with set: keep only elements present in target_set
intersection_counter = Counter({k: v for k, v in counter.items() if k in target_set})
# Result: Counter({1:2, 4:1, 5:1})

# Convert back to list (preserving duplicates
result_list = list(intersection_counter.elements())
print(result_list)   # [1, 1, 4, 5]
# Length of that list
print(len(result_list))  # 4
```

```python
from typing import List,Set
from collections import Counter
def final_score(input_array:List[int],A:Set[int],B:Set[int])->int:
    
    counter = Counter(input_array)
    intersection_counter_A = Counter({k: v for k, v in counter.items() if k in A})
    intersection_counter_B = Counter({k: v for k, v in counter.items() if k in B})
    result_list_A = list(intersection_counter_A.elements())
    result_list_B = list(intersection_counter_B.elements())

    return len(result_list_A) - len(result_list_B)
```
