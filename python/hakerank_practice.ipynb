```python
# input: 1 1 1 2
x,y,z,n = map(int,input().split())
final_list = []
for i in range(x+1):
    for j in range(y+1):
        for k in range(z+1):
            if i+j+k != n:
                final_list.append([i,j,k])
print(final_list)

```

```python
N = int(input())

for i in range(N):
    command,*variables = input().split()
    print(command,variables) 
```

```python
"""
insert i e: Insert integer  at position .
print: Print the list.
remove e: Delete the first occurrence of integer .
append e: Insert integer  at the end of the list.
sort: Sort the list.
pop: Pop the last element from the list.
reverse: Reverse the list."""

N = int(input())
sample_list = []
for i in range(N):
    command,*variables = input().split()
    if command == "insert":
        sample_list.insert(int(variables[0]),int(variables[1]))
    elif command == "print":
        print(sample_list)
    elif command == "remove":
        sample_list.remove(int(variables[0]))
    elif command == "append":
        sample_list.append(int(variables[0]))
    elif command == "sort":
        sample_list.sort()
    elif command == "pop":
        sample_list.pop()
    elif command == "reverse":
        sample_list.reverse()
    else:
        print("wrong command")                
```

```python
name_list = []
score_list = []

for _ in range(int(input())):
    name = input()
    score = float(input())
    name_list.append(name)
    score_list.append(score)

print(name_list,"----", score_list)    

second_lowest_scores = sorted(list(set(score_list)))[1]
print("sec_low_sco",second_lowest_scores)

second_lowest_scores_indexes = [index for index,value in enumerate(score_list) if value == second_lowest_scores]
print("second_lowest_scores_indexes:",second_lowest_scores_indexes)

names_with_second_highest_scores = [name_list[i] for i in second_lowest_scores_indexes]
print("names_with_second_highest_scores",names_with_second_highest_scores)
print(*sorted(names_with_second_highest_scores),sep="\n")

```
