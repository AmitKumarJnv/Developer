# Accessing and modifying data

## 1. The traditional way: make the data private and use getter and setters:

```python
class User:
    def __init__(self,username,email:str,password):
        self.username = username
        self._email = email
        self.password = password

    def clean_email(self):
        return self._email.lower().strip()   
        
    
```

```python
user1 = User("AMit", "Amit123@gmail.com    ","123")
```

```python
user1.clean_email()
```

```python
user1._email
```

```python
user1._email=user1.clean_email()
```

```python
user1._email
```

Private vs protected variable

Protected 
- _email
- accessible within and outside the class, but its developers responsibility not to use the protected variable outside the class

Private
- __email
- Accessible only within the class

```python
from datetime import datetime

class Users:
    def __init__(self,name,email,mobile_number):
        self.name = name
        self._email = email
        self.mobile_number = mobile_number

    def get_email(self):
        print(f"email is accessed at {datetime.now()}")
        return self._email

    def set_email(self,new_email):
        if "@" in new_email:
            self._email = new_email
        else:
            print("Invalid email")    



        
```

```python
user1 = Users("Amit","abc@gmail.com","111222333")
```

```python
user1.get_email()
```

```python
user1.set_email("123")
```

```python
user1.get_email()
```

```python
from datetime import datetime

class Users:
    def __init__(self,name,email,mobile_number):
        self.name = name
        self.__email = email
        self.mobile_number = mobile_number

    @property
    def email(self):
        print(f"email is accessed at {datetime.now()}")
        return self.__email

    @email.setter
    def email(self,new_email):
        if "@" in new_email:
            self._email = new_email
        else:
            print("Invalid email")    
```

```python
user1 = Users("Amit","jnvamit1666@gmail.com","1234123123") 
```

```python
user1.email
```

```python
user1.email = "abc"
```

<b> Static Attributes </b>
- A static attribute (sometimes called as a class attribute) is an attribute that belongs to the class itself, not to any specific instance of the class

```python
class Users:
    user_counts = 0

    def __init__(self,name,email):
        self.name = name 
        self.email = email
        Users.user_counts += 1

    def user_info(self):
        print(f"My name is {self.name} and my email is {self.email}")    
    
        
```

```python
user1 = Users("Amit","amita2123")
user2 = Users("Anand","jnv@123")
```

```python
Users.user_counts
```

```python
user1.user_counts
```
