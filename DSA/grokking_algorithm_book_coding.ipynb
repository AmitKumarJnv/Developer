## Binary Serach

```python
## Binary Search

from typing import List, Optional, Union

def binary_search(input_sorted_list:List[Union[int,float]], search_term:Union[int,float])-> Optional[Union[int,float]]:
    low = 0
    high = len(input_sorted_list) - 1

    while low <= high:
        mid = (low + high) // 2
        if input_sorted_list[mid] == search_term:
            return mid
        elif input_sorted_list[mid] > search_term:
            high = mid - 1
        else:
            low = mid + 1

    return None            
```

```python
my_list = [1.0,3.0,5.0,7.0,9.0]

print(binary_search(my_list,3.0)) # 1
print(binary_search(my_list,-1.0)) # None
```

## Running Time

#### - If the maximum number of guesses is the same as the size of the list, it is known as linear time.
#### - Binary Search runs in logarithmic time.

## Big O Notation

#### - Big O notation is special notation that tells you how fast an algorithm is.
#### - Big O notation lets you compare the number of operations. It tells you how fast the algorithm grows.
#### - Binary Search needs log n operations to check a list of size n. In Big O notation, its O(log n).
#### - Big O notation is written as O(n), where n is `  number of operations` 
#### - Along with worst-case run time, its also important to look at the average-case run time.\n
#### - `Some common Big O run times`
- `O(log n)`: also known as log time, i.e Binary Search.
- `O(n)`: also known as linear time, i.e Simple Search
- `O(n * log n)`: A fast sorting algorithm, like quicksort
- `O(n^2)`: A slow sorting algorithm, like selection sort
- `O(n!)`: A slow algorithm , i.e traveling salesperson. \n\n


# Selection Sort

#### - `Randon & Sequential Access`:
- Random is used by arrays
- Sequential is used by linked lists


```python
def smallest_index(input_list:List[Union[int,float]])->Union[int,float]:
    smallest_value = input_list[0]
    smallest_index_value = 0
    for i in range(1,len(input_list)):
        if input_list[i] < smallest_value:
            smallest_value = input_list[i]
            smallest_index_value = i

    return smallest_index_value        
```

```python
print(smallest_index([2,5,8,-9,4])) # 3
print(smallest_index([2,5,-9,8,4])) # 2
print(smallest_index([2,-9,8,5,4])) # 1
```

```python
def selection_sort(unsorted_list:List[Union[int,float]])->List[Union[int,float]]:
    sorted_list = list()
    for i in range(len(unsorted_list)):
        smallest_index_value = smallest_index(unsorted_list)
        sorted_list.append(unsorted_list.pop(smallest_index_value))
    return sorted_list     
```

```python
print(selection_sort([2,5,8,-9,4]))
print(selection_sort([2,5,-9,8,4]))
print(selection_sort([2,-9,8,5,4]))
```
